```python
# ==========================================
# 📡 Download Flood + Wildfire Data from Firebase
# ==========================================

!pip install firebase-admin pandas --quiet

import firebase_admin
from firebase_admin import credentials, db, get_app, initialize_app
import pandas as pd
from google.colab import files

# -------------------------------
# 🔐 Firebase Initialization
# -------------------------------
try:
    app = get_app()
    print("✅ Firebase already initialized.")
except ValueError:
    cred = credentials.Certificate(
        "earlywarningsystem-53511-firebase-adminsdk-fbsvc-21b7d41508.json"
    )
    initialize_app(cred, {
        "databaseURL": "https://earlywarningsystem-53511-default-rtdb.firebaseio.com/"
    })
    print("🚀 Firebase initialized successfully.")

# ============================================================
# FUNCTION: Download any node as CSV
# ============================================================
def download_node_as_csv(node_path, filename, column_order=None):
    print(f"\n📡 Fetching data from /{node_path} ...")

    ref = db.reference(node_path)
    data = ref.get()

    if not data:
        print(f"⚠️ No data found in /{node_path}")
        return

    df = pd.DataFrame.from_dict(data, orient="index")

    # Optional column ordering
    if column_order:
        for col in column_order:
            if col not in df.columns:
                df[col] = None
        df = df[column_order]

    # Cleanup boolean values
    for col in df.columns:
        if "detected" in col:
            df[col] = df[col].map({1: "Yes", 0: "No", True: "Yes", False: "No"})

    df.drop_duplicates(inplace=True)

    # Save
    df.to_csv(filename, index=False)
    print(f"✅ Saved: {filename}")

    # Preview
    display(df.tail())

    # Auto-download
    files.download(filename)


# ============================================================
# 🌊 FLOOD DATA EXPORT
# ============================================================
download_node_as_csv(
    node_path="floodData",
    filename="flood_dataset.csv",
    column_order=[
        "timestamp",
        "water_level_cm",
        "rain_intensity_percent",
        "water_sensor_percent",
        "flood_detected",
        "risk_level",
    ]
)

# ============================================================
# 🔥 WILDFIRE DATA EXPORT
# ============================================================
download_node_as_csv(
    node_path="wildfireData",
    filename="wildfire_dataset.csv",
    column_order=[
        "timestamp",
        "temperature",
        "humidity",
        "gas_value",
        "flame_detected",
        "wildfire_detected",
        "risk_level",
    ]
)



# ============================================================
# 🌍 EARTHQUAKE DATA EXPORT
# ============================================================
download_node_as_csv(
    node_path="earthquakeData",
    filename="earthquake_dataset.csv",
    column_order=[
        "timestamp",
        "motion",
        "vibration_detected",
        "earthquake_detected",
        "risk_level",
    ]
)

# ============================================================
# ⛰️ LANDSLIDE DATA EXPORT
# ============================================================
download_node_as_csv(
    node_path="landslideData",
    filename="landslide_dataset.csv",
    column_order=[
        "timestamp",
        "pressure",
        "soil_moisture",
        "landslide_detected",
        "risk_level",
    ]
)


# ============================================================
# 🌪️ STORM PREDICTION (FUSION DATA) EXPORT
# ============================================================
download_node_as_csv(
    node_path="fusionData",
    filename="storm_prediction_dataset.csv",
    column_order=[
        "timestamp",
        "distance_cm",
        "rain_percent",
        "water_percent",
        "wind_speed_mps",
        "flood_risk",
        "storm_risk",
        "fused_event",
        "fused_risk",
    ]
)

print("\n✅ DONE: All datasets downloaded successfully.")



```
